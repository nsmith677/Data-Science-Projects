# Introduction

The intention of this project is to explore visualizations using pandas. The data set we will be using is on the job outcomes of students who graduated from college between 2010 and 2012. The data originally was released by American Community Survey. FiveThirtyEight cleaned the dataset and it can be found on their [github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
```

```python
#drop rows containing missing values
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
```
